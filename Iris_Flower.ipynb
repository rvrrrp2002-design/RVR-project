```python
 import numpy as np
 import pandas as pd
 import matplotlib.pyplot as plt
 import seaborn as sns
 import plotly.express as px


```

```python
df=pd.read_csv("/content/IRIS.csv")
df.head(10)
```

Data Pre-processing

```python
#importing libraris used for encoding
from sklearn.preprocessing import LabelEncoder
from pandas.core.dtypes.common import is_numeric_dtype
le=LabelEncoder()
```

```python
for i in df.columns:
  if is_numeric_dtype(df[i]):
    continue
  else:
    df[i]=le.fit_transform(df[i])
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
#Assigning the predictor and target variable
x=df.drop('species',axis=1)
y=df['species']
```

```python

X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Model building

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
print('score for the train data',model.score(X_train,y_train))
print('score for the test data',model.score(X_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#Building a machine learning model
model1=RandomForestClassifier(n_estimators= 300, random_state=100)
model1.fit(X_train,y_train)
```

```python
# evaluating and verifying accuracy
print('score for the train data',model1.score(X_train,y_train))
print('score for the test data',model1.score(X_test,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model2=KNeighborsClassifier()
model2.fit(X_train,y_train)
```

```python
#evaluating and verifying accuracy
print('score for the train data',model2.score(X_train,y_train))
print('score for the test data',model2.score(X_test,y_test))
```

Analzing the score of the 3 process the Random forest Classifier has higher effiency in train and test data
